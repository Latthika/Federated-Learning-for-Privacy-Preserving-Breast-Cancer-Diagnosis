```python
# ============================================================
# FULL 1D-CNN  Federated Experiments + Statistical Tests
# ============================================================

# If needed in a fresh runtime:
# !pip install tensorflow --quiet

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, classification_report
)
from sklearn.feature_selection import SelectKBest, f_classif

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers

# Statistics
from scipy.stats import ttest_rel, wilcoxon, friedmanchisquare

sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 120

# ----------------------------
# GLOBAL CONFIG
# ----------------------------
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

N_CLIENTS = 10
N_TRIALS = 50

CENTRAL_EPOCHS = 30
LOCAL_EPOCHS   = 20
BATCH_SIZE     = 32
LEARNING_RATE  = 1e-3

GLOBAL_DIST_SHIFT_STD = 0.3   # one-time global noise on train set
DP_AGG_NOISE_STD      = 0.10  # DP noise at aggregation (proba-level)

FS_K           = 20   # number of features kept for training
EDA_TOP_FEATS  = 6    # number of features to visualize in some plots

FEATURE_SHIFT_CLIENTS = [2, 7]  # simulate client drift
FEATURE_SHIFT_DELTA   = 0.5     # simple additive feature shift

# ============================================================
# 1. LOAD + STANDARDIZE DATA
# ============================================================
def load_and_preprocess_wdbc():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points',
            'symmetry', 'fractal_dimension'
        ]
    ]

    df = pd.read_csv(url, header=None, names=columns)
    df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1})
    return df

df_eda = load_and_preprocess_wdbc()

# Basic EDA prints
print("\nDataset shape:", df_eda.shape)
print("\nClass counts:")
print(df_eda['diagnosis'].value_counts())
print("\nClass proportions:")
print(df_eda['diagnosis'].value_counts(normalize=True).round(4))

# ============================================================
# 2. EDA (histograms, boxplots, small correlation)
# ============================================================
df_eda['diagnosis_label'] = df_eda['diagnosis'].map({0: 'Benign', 1: 'Malignant'})

# show a few histograms for quick checks
sample_feats = df_eda.columns[2:2+EDA_TOP_FEATS]  # skip id, diagnosis
for feat in sample_feats:
    plt.figure(figsize=(5,3))
    plt.hist(df_eda[feat], bins=25, alpha=0.7)
    plt.title(f"Histogram: {feat}")
    plt.tight_layout()
    plt.show()

# boxplots
plt.figure(figsize=(8,4))
df_eda.iloc[:, 2:2+EDA_TOP_FEATS].boxplot(rot=45)
plt.title("Boxplots of some features")
plt.tight_layout()
plt.show()

# quick correlation heatmap of top few features
corr = df_eda.iloc[:, 2:2+EDA_TOP_FEATS].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation heatmap (sample features)")
plt.tight_layout()
plt.show()

# ============================================================
# 3. TRAINTEST SPLIT + ONE-TIME GLOBAL NOISE
# ============================================================
X_all = df_eda.drop(columns=['id', 'diagnosis', 'diagnosis_label']).values.astype(np.float32)
y_all = df_eda['diagnosis'].values.astype(np.int64)

scaler = StandardScaler()
X_scaled_all = scaler.fit_transform(X_all)

X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_scaled_all, y_all, test_size=0.2, stratify=y_all, random_state=RANDOM_SEED
)

print("\nTrain/test sizes:")
print(len(y_train), len(y_test))

# One-time global shift on training set
X_train_full = X_train_full + np.random.normal(0, GLOBAL_DIST_SHIFT_STD, X_train_full.shape)
print("Applied one-time Gaussian shift to training pool.")

# ============================================================
# 4. FEATURE IMPORTANCE + SelectKBest
# ============================================================
f_scores, _ = f_classif(X_train_full, y_train)
feature_names = df_eda.columns[2:-1].tolist()  # all features excluding id, diagnosis, diagnosis_label
sorted_idx = np.argsort(f_scores)[::-1]
top_idx = sorted_idx[:EDA_TOP_FEATS]
print("\nTop features by ANOVA F-score:")
for idx in top_idx:
    print(f"  {feature_names[idx]}: F = {f_scores[idx]:.3f}")

selector = SelectKBest(score_func=f_classif, k=FS_K)
X_train = selector.fit_transform(X_train_full, y_train)
X_test  = selector.transform(X_test_full)
input_dim = X_train.shape[1]
print(f"\nSelected top {FS_K} features. Input dim = {input_dim}")

# ============================================================
# 5. MANUAL CLIENT SPLIT (LOCKED ONCE)
# ============================================================
MANUAL_LABEL_ALLOC = [
    (33, 19), (10, 2), (61,40), (34,31), (23,11),
    (45,21), (33,18), (26,20), (13,5), (7,3)
]
client_totals = [b+m for (b,m) in MANUAL_LABEL_ALLOC]
assert sum(client_totals) == len(y_train), "Client sizes must sum to train set size!"

benign_idx = np.where(y_train == 0)[0].tolist()
malig_idx  = np.where(y_train == 1)[0].tolist()
random.shuffle(benign_idx)
random.shuffle(malig_idx)

client_indices = []
sb = sm = 0
for (b_ct, m_ct) in MANUAL_LABEL_ALLOC:
    idx = benign_idx[sb:sb+b_ct] + malig_idx[sm:sm+m_ct]
    random.shuffle(idx)
    client_indices.append(idx)
    sb += b_ct
    sm += m_ct

client_splits = [(X_train[idx], y_train[idx]) for idx in client_indices]
print("\nClient splits (locked). Per-client counts:")
for cid, (_, yi) in enumerate(client_splits):
    total = len(yi)
    b = int(np.sum(yi==0))
    m = int(np.sum(yi==1))
    print(f"  Client {cid}: total={total}, benign={b} ({b/total:.1%}), malignant={m} ({m/total:.1%})")

# ============================================================
# 6. CLIENT DRIFT VISUALIZATION
# ============================================================
global_mean = X_train.mean(axis=0)
drifts = [np.linalg.norm(Xi.mean(axis=0) - global_mean) for Xi,_ in client_splits]
plt.figure(figsize=(7,4))
plt.plot(range(N_CLIENTS), drifts, marker='o')
plt.xlabel("Client ID"); plt.ylabel("L2 distance to global mean")
plt.title("Client drift (feature mean distance)")
plt.grid(alpha=0.3); plt.tight_layout(); plt.show()

# ============================================================
# 7. 1D-CNN MODEL & HELPERS
# ============================================================
def build_cnn_model(input_dim):
    model = models.Sequential([
        layers.Input(shape=(input_dim,1)),
        layers.Conv1D(32, kernel_size=3, activation='relu', padding='same'),
        layers.MaxPooling1D(2),
        layers.Conv1D(64, kernel_size=3, activation='relu', padding='same'),
        layers.GlobalAveragePooling1D(),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(
        optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

def train_cnn_local(X_local, y_local, seed, epochs):
    tf.random.set_seed(seed)
    X_loc = X_local.reshape(-1, input_dim, 1)
    m = build_cnn_model(input_dim)
    m.fit(X_loc, y_local, epochs=epochs, batch_size=BATCH_SIZE, shuffle=True, verbose=0)
    return m

def ensemble_predict(models, sizes, X_eval, dp=False):
    total = float(sum(sizes))
    X_e = X_eval.reshape(-1, input_dim, 1)
    proba = np.zeros((X_e.shape[0],), dtype=float)
    for m, n_k in zip(models, sizes):
        w = n_k/total
        proba += w * m.predict(X_e, verbose=0).ravel()
    if dp:
        proba += np.random.normal(0, DP_AGG_NOISE_STD, size=proba.shape)
        proba = np.clip(proba, 0.0, 1.0)
    return (proba >= 0.5).astype(int)

def score_vector(y_true, y_pred):
    return (
        accuracy_score(y_true, y_pred),
        precision_score(y_true, y_pred, average='macro', zero_division=0),
        recall_score(y_true, y_pred, average='macro', zero_division=0),
        f1_score(y_true, y_pred, average='macro', zero_division=0)
    )

# ============================================================
# 8. 5 EXPERIMENTS  N_TRIALS
# ============================================================
metrics_exp1 = []  # Central
metrics_exp2 = []  # FedAvg
metrics_exp3 = []  # FedAvg + DP
metrics_exp4 = []  # FedProx
metrics_exp5 = []  # FedProx + DP

print("\n========== RUNNING 5 EXPERIMENTS ", N_TRIALS, "TRIALS ==========\n")
for t in range(1, N_TRIALS+1):
    print(f"--- TRIAL {t}/{N_TRIALS} ---")

    # 1. Central CNN baseline
    central_model = train_cnn_local(X_train, y_train, seed=RANDOM_SEED + 1000*t, epochs=CENTRAL_EPOCHS)
    y1 = (central_model.predict(X_test.reshape(-1,input_dim,1), verbose=0).ravel() >= 0.5).astype(int)
    m1 = score_vector(y_test, y1); metrics_exp1.append(m1)
    print(f"1 Central -> Acc={m1[0]:.4f}, F1={m1[3]:.4f}")

    # 2. FedAvg CNN (train local models, aggregate probs)
    models2, sizes2 = [], []
    for cid, (Xi, yi) in enumerate(client_splits):
        X_loc = Xi.copy()
        if cid in FEATURE_SHIFT_CLIENTS:
            X_loc = X_loc + FEATURE_SHIFT_DELTA
        mdl = train_cnn_local(X_loc, yi, seed=RANDOM_SEED + 2000*t + cid, epochs=LOCAL_EPOCHS)
        models2.append(mdl); sizes2.append(len(yi))
    y2 = ensemble_predict(models2, sizes2, X_test, dp=False)
    m2 = score_vector(y_test, y2); metrics_exp2.append(m2)
    print(f"2 FedAvg -> Acc={m2[0]:.4f}, F1={m2[3]:.4f}")

    # 3. FedAvg + DP
    y3 = ensemble_predict(models2, sizes2, X_test, dp=True)
    m3 = score_vector(y_test, y3); metrics_exp3.append(m3)
    print(f"3 FedAvg+DP -> Acc={m3[0]:.4f}, F1={m3[3]:.4f}")

    # 4. FedProx (conceptual)  local models (could include prox term in real FL)
    models4, sizes4 = [], []
    for cid, (Xi, yi) in enumerate(client_splits):
        X_loc = Xi.copy()
        if cid in FEATURE_SHIFT_CLIENTS:
            X_loc = X_loc + FEATURE_SHIFT_DELTA
        mdl = train_cnn_local(X_loc, yi, seed=RANDOM_SEED + 3000*t + cid, epochs=LOCAL_EPOCHS)
        models4.append(mdl); sizes4.append(len(yi))
    y4 = ensemble_predict(models4, sizes4, X_test, dp=False)
    m4 = score_vector(y_test, y4); metrics_exp4.append(m4)
    print(f"4 FedProx -> Acc={m4[0]:.4f}, F1={m4[3]:.4f}")

    # 5. FedProx + DP
    y5 = ensemble_predict(models4, sizes4, X_test, dp=True)
    m5 = score_vector(y_test, y5); metrics_exp5.append(m5)
    print(f"5 FedProx+DP -> Acc={m5[0]:.4f}, F1={m5[3]:.4f}\n")

print("\nAll trials finished.\n")

# ============================================================
# 9. METRIC SUMMARY (MEAN  STD)
# ============================================================
cols = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']
df1 = pd.DataFrame(metrics_exp1, columns=cols)
df2 = pd.DataFrame(metrics_exp2, columns=cols)
df3 = pd.DataFrame(metrics_exp3, columns=cols)
df4 = pd.DataFrame(metrics_exp4, columns=cols)
df5 = pd.DataFrame(metrics_exp5, columns=cols)

summary_df = pd.DataFrame({
    "experiment": [
        "Central CNN Baseline",
        "FedAvg CNN",
        "FedAvg CNN + DP",
        "FedProx CNN",
        "FedProx CNN + DP"
    ],
    "accuracy_mean":  [df['accuracy'].mean() for df in [df1, df2, df3, df4, df5]],
    "accuracy_std":   [df['accuracy'].std()  for df in [df1, df2, df3, df4, df5]],
    "f1_mean":        [df['f1_macro'].mean()  for df in [df1, df2, df3, df4, df5]],
    "f1_std":         [df['f1_macro'].std()   for df in [df1, df2, df3, df4, df5]],
    "prec_mean":      [df['precision_macro'].mean() for df in [df1, df2, df3, df4, df5]],
    "recall_mean":    [df['recall_macro'].mean()    for df in [df1, df2, df3, df4, df5]],
})

print("===== 50-TRIAL SUMMARY (MEAN  STD) =====")
print(summary_df.round(4).to_string(index=False))

# ============================================================
# 10. PLOTS: mean performance and comparisons
# ============================================================
plt.figure(figsize=(9,5))
plt.plot(summary_df['experiment'], summary_df['accuracy_mean'], marker='o')
plt.title("1D-CNN Accuracy (mean over trials)")
plt.xticks(rotation=20); plt.grid(alpha=0.3); plt.show()

plt.figure(figsize=(9,5))
plt.plot(summary_df['experiment'], summary_df['f1_mean'], marker='o')
plt.title("1D-CNN Macro F1 (mean over trials)")
plt.xticks(rotation=20); plt.grid(alpha=0.3); plt.show()

# ============================================================
# 11. STATISTICAL TESTING (Paired t-test, Wilcoxon, Friedman)
# ============================================================
# Organize matrices
acc_matrix = np.vstack([df1['accuracy'].values, df2['accuracy'].values, df3['accuracy'].values, df4['accuracy'].values, df5['accuracy'].values])
f1_matrix  = np.vstack([df1['f1_macro'].values, df2['f1_macro'].values, df3['f1_macro'].values, df4['f1_macro'].values, df5['f1_macro'].values])

exp_names = [
    "Central CNN",
    "FedAvg CNN",
    "FedAvg CNN + DP",
    "FedProx CNN",
    "FedProx CNN + DP"
]

# Pairwise index list
import itertools
pairs = list(itertools.combinations(range(5), 2))
n_pairs = len(pairs)

print("\n=== Pairwise Paired t-tests (Accuracy) ===")
ttest_results_acc = {}
for i,j in pairs:
    t_stat, p = ttest_rel(acc_matrix[i], acc_matrix[j])
    ttest_results_acc[(i,j)] = p
    print(f"{exp_names[i]} vs {exp_names[j]}: p = {p:.6e}")

print("\n=== Pairwise Paired t-tests (F1) ===")
ttest_results_f1 = {}
for i,j in pairs:
    t_stat, p = ttest_rel(f1_matrix[i], f1_matrix[j])
    ttest_results_f1[(i,j)] = p
    print(f"{exp_names[i]} vs {exp_names[j]}: p = {p:.6e}")

print("\n=== Wilcoxon signed-rank tests (Accuracy) ===")
wilc_results_acc = {}
for i,j in pairs:
    try:
        stat, p = wilcoxon(acc_matrix[i], acc_matrix[j])
    except Exception as e:
        p = np.nan
    wilc_results_acc[(i,j)] = p
    print(f"{exp_names[i]} vs {exp_names[j]}: p = {p if not np.isnan(p) else 'nan'}")

print("\n=== Wilcoxon signed-rank tests (F1) ===")
wilc_results_f1 = {}
for i,j in pairs:
    try:
        stat, p = wilcoxon(f1_matrix[i], f1_matrix[j])
    except Exception as e:
        p = np.nan
    wilc_results_f1[(i,j)] = p
    print(f"{exp_names[i]} vs {exp_names[j]}: p = {p if not np.isnan(p) else 'nan'}")

# Friedman tests
fried_acc = friedmanchisquare(*[acc_matrix[k] for k in range(5)])
fried_f1  = friedmanchisquare(*[f1_matrix[k] for k in range(5)])
print("\n=== Friedman tests ===")
print(f"Accuracy: statistic={fried_acc.statistic:.4f}, p={fried_acc.pvalue:.6e}")
print(f"F1:       statistic={fried_f1.statistic:.4f}, p={fried_f1.pvalue:.6e}")

# Bonferroni adjustment
def bonferroni_adjust(pdict):
    adjusted = {}
    for k,p in pdict.items():
        if p is None or np.isnan(p):
            adjusted[k] = np.nan
        else:
            adjusted[k] = min(1.0, p * n_pairs)
    return adjusted

ttest_acc_bonf = bonferroni_adjust(ttest_results_acc)
ttest_f1_bonf  = bonferroni_adjust(ttest_results_f1)
wilc_acc_bonf  = bonferroni_adjust(wilc_results_acc)
wilc_f1_bonf   = bonferroni_adjust(wilc_results_f1)

print("\n=== Bonferroni-adjusted paired t-test p-values (accuracy) ===")
for (i,j), p in ttest_acc_bonf.items():
    print(f"{exp_names[i]} vs {exp_names[j]}: p_adj = {p:.6e}")

# ============================================================
# 12. P-VALUE MATRICES, HEATMAPS, BOXPLOTS
# ============================================================
def make_pval_matrix(pdict):
    mat = np.ones((5,5))
    for (i,j), p in pdict.items():
        mat[i,j] = p if not np.isnan(p) else np.nan
        mat[j,i] = mat[i,j]
    np.fill_diagonal(mat, 0.0)
    return mat

p_ttest_acc_mat = make_pval_matrix(ttest_results_acc)
p_wilc_acc_mat  = make_pval_matrix(wilc_results_acc)
p_ttest_f1_mat  = make_pval_matrix(ttest_results_f1)
p_wilc_f1_mat   = make_pval_matrix(wilc_results_f1)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.heatmap(p_ttest_acc_mat, annot=True, fmt=".2e", xticklabels=exp_names, yticklabels=exp_names, cmap="viridis")
plt.title("Paired t-test p-values (accuracy)")
plt.subplot(1,2,2)
sns.heatmap(p_wilc_acc_mat, annot=True, fmt=".2e", xticklabels=exp_names, yticklabels=exp_names, cmap="viridis")
plt.title("Wilcoxon p-values (accuracy)")
plt.tight_layout()
plt.show()

# Boxplots
acc_df_plot = pd.DataFrame({
    exp_names[0]: acc_matrix[0],
    exp_names[1]: acc_matrix[1],
    exp_names[2]: acc_matrix[2],
    exp_names[3]: acc_matrix[3],
    exp_names[4]: acc_matrix[4],
})
plt.figure(figsize=(9,5))
sns.boxplot(data=acc_df_plot)
plt.title("Accuracy distributions across experiments")
plt.xticks(rotation=20); plt.tight_layout(); plt.show()

f1_df_plot = pd.DataFrame({
    exp_names[0]: f1_matrix[0],
    exp_names[1]: f1_matrix[1],
    exp_names[2]: f1_matrix[2],
    exp_names[3]: f1_matrix[3],
    exp_names[4]: f1_matrix[4],
})
plt.figure(figsize=(9,5))
sns.boxplot(data=f1_df_plot)
plt.title("Macro F1 distributions across experiments")
plt.xticks(rotation=20); plt.tight_layout(); plt.show()

# ============================================================
# 13. FINAL PUBLICATION TABLE: meanstd & significance vs Central
# ============================================================
alpha = 0.05
vs_central = []
for k in range(5):
    mean_acc = acc_matrix[k].mean()
    std_acc  = acc_matrix[k].std()
    mean_f1  = f1_matrix[k].mean()
    std_f1   = f1_matrix[k].std()
    if k == 0:
        t_p = np.nan
        w_p = np.nan
        sig = ""
    else:
        # pair (0,k) if found else (k,0)
        key = (0,k) if (0,k) in ttest_results_acc else (k,0)
        t_p = ttest_results_acc.get(key, np.nan)
        w_p = wilc_results_acc.get(key, np.nan)
        sig = "*" if (not np.isnan(t_p) and t_p < alpha) or (not np.isnan(w_p) and w_p < alpha) else ""
    vs_central.append({
        "experiment": exp_names[k],
        "acc_mean": mean_acc, "acc_std": std_acc,
        "f1_mean": mean_f1, "f1_std": std_f1,
        "sig_vs_central": sig,
        "t_p_raw": t_p, "wilc_p_raw": w_p
    })

final_df = pd.DataFrame(vs_central)
final_df_display = final_df.copy()
final_df_display['accuracy'] = final_df_display.apply(lambda r: f"{r['acc_mean']:.4f}  {r['acc_std']:.4f}{r['sig_vs_central']}", axis=1)
final_df_display['f1'] = final_df_display.apply(lambda r: f"{r['f1_mean']:.4f}  {r['f1_std']:.4f}{r['sig_vs_central']}", axis=1)
final_df_display = final_df_display[['experiment','accuracy','f1','t_p_raw','wilc_p_raw']]
print("\n===== FINAL RESULTS TABLE (mean  std)  '*' indicates significant vs Central (p<0.05) =====\n")
print(final_df_display.to_string(index=False))

# Done
print("\n=== STATISTICAL ANALYSIS COMPLETE ===")
```
