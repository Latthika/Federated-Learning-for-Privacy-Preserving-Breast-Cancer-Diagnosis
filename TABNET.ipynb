```python
# ============================================================
# TABNET + FEDERATED LEARNING + DP — PART 1
# (Installs, imports, EDA, scaling, client split, helpers)
# ============================================================

# If running in a fresh Colab runtime, uncomment the install line below.
# (Usually pytorch-tabnet is available; leave commented if already installed.)
#!pip install pytorch-tabnet --quiet

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.feature_selection import f_classif

# TabNet
from pytorch_tabnet.tab_model import TabNetClassifier
import torch

# Statistical tests imports reserved for Part 3 but safe to import now
from scipy.stats import ttest_rel, wilcoxon, friedmanchisquare

sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 120

# ----------------------------
# GLOBAL CONFIG
# ----------------------------
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)

N_CLIENTS = 10
N_TRIALS = 50

GLOBAL_DIST_SHIFT_STD = 0.3   # one-time global noise on train data
DP_AGG_NOISE_STD = 0.10       # DP noise at aggregation

FEATURE_SHIFT_CLIENTS = [2, 7]  # clients with drift
FEATURE_SHIFT_DELTA = 0.5       # simple feature shift

# TabNet training config
MAX_EPOCHS = 50
PATIENCE = 10

TABNET_PARAMS = dict(
    n_d=16,
    n_a=16,
    n_steps=4,
    n_independent=2,
    n_shared=2,
    lambda_sparse=1e-4,
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=1e-3),
    mask_type='sparsemax',
    verbose=0,
)

# ============================================================
# 1. LOAD DATA + BASIC EDA
# ============================================================
def load_wdbc_raw():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points',
            'symmetry', 'fractal_dimension'
        ]
    ]
    df = pd.read_csv(url, header=None, names=columns)
    df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1})
    return df

df_raw = load_wdbc_raw()
print("Data shape:", df_raw.shape)
print(df_raw.head(), "\n")

print("Class distribution (counts):")
print(df_raw['diagnosis'].value_counts().rename({0: "Benign (0)", 1: "Malignant (1)"}))
print()

feature_cols = [c for c in df_raw.columns if c not in ['id', 'diagnosis']]
X_raw = df_raw[feature_cols].copy()
y_raw = df_raw['diagnosis'].values

print("Feature summary statistics (first 6 rows):")
print(X_raw.describe().T.head(6))
print("\n")

# ============================================================
# 2. EDA PLOTS (brief, safe for notebook interactive)
# ============================================================
important_demo_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
df_plot = df_raw[important_demo_features + ['diagnosis']].copy()
df_plot['diagnosis_str'] = df_plot['diagnosis'].map({0: 'Benign', 1: 'Malignant'})

for feat in important_demo_features:
    plt.figure(figsize=(6, 3.5))
    sns.histplot(data=df_plot, x=feat, hue="diagnosis_str", kde=True, stat="density", common_norm=False)
    plt.title(f"Distribution of {feat} by Class")
    plt.tight_layout()
    plt.show()

# Boxplots for demo features
plt.figure(figsize=(9, 4))
melted = df_plot.melt(id_vars=['diagnosis_str'], value_vars=important_demo_features,
                      var_name='feature', value_name='value')
sns.boxplot(data=melted, x='feature', y='value', hue='diagnosis_str')
plt.title("Boxplots of Selected Features by Class")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

# ============================================================
# 3. ANOVA (feature ranking) + Correlation heatmap (top features)
# ============================================================
F_vals, p_vals = f_classif(X_raw.values, y_raw)
anova_df = pd.DataFrame({
    'feature': feature_cols,
    'f_score': F_vals,
    'p_value': p_vals
}).sort_values('f_score', ascending=False)

top_k = 10
top_features = anova_df.head(top_k)['feature'].tolist()
print("\nTop 10 features by ANOVA F-score:")
print(anova_df.head(10))
print("\n")

# Correlation heatmap for top features
plt.figure(figsize=(9, 7))
corr_top = df_raw[top_features].corr()
sns.heatmap(
    corr_top,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    cbar_kws={"shrink": 0.8},
    annot_kws={"size": 8}
)
plt.title("Correlation Heatmap (Top 10 ANOVA Features)")
plt.tight_layout()
plt.show()

# Pairplot for top 4 features + class (smaller, faster)
pair_feats = top_features[:4]
df_pair = df_raw[pair_feats + ['diagnosis']].copy()
df_pair['diagnosis_str'] = df_pair['diagnosis'].map({0: 'Benign', 1: 'Malignant'})

sns.pairplot(
    data=df_pair,
    vars=pair_feats,
    hue='diagnosis_str',
    diag_kind='kde',
    plot_kws={"alpha": 0.6, "s": 20, "edgecolor": "none"},
    height=1.6
)
plt.suptitle("Pairplot of Top 4 Features (by ANOVA F-score)", y=1.02)
plt.show()

# ============================================================
# 4. STANDARD SCALING + TRAIN/TEST SPLIT
# ============================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw.values)
y = y_raw.copy()

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=RANDOM_SEED
)

print("Central Train/Test Split Size:")
print(f"  Train samples: {len(y_train)}")
print(f"  Test samples : {len(y_test)}\n")

# One-time global noise for distribution shift
print(f"Applying one-time Gaussian noise to training data (std={GLOBAL_DIST_SHIFT_STD})...")
X_train = X_train + np.random.normal(0, GLOBAL_DIST_SHIFT_STD, X_train.shape)
print("Noise added to training pool.\n")

# ============================================================
# 5. MANUAL CLIENT SPLIT (LOCKED)
# ============================================================
MANUAL_LABEL_ALLOC = [
    (33, 19),  # Client 0: 52
    (10,  2),  # Client 1: 12
    (61, 40),  # Client 2: 101
    (34, 31),  # Client 3: 65
    (23, 11),  # Client 4: 34
    (45, 21),  # Client 5: 66
    (33, 18),  # Client 6: 51
    (26, 20),  # Client 7: 46
    (13,  5),  # Client 8: 18
    ( 7,  3),  # Client 9: 10
]

client_totals = [b + m for (b, m) in MANUAL_LABEL_ALLOC]
assert sum(client_totals) == len(y_train), "Client sizes must sum to train set size!"

benign_idx = np.where(y_train == 0)[0].tolist()
malig_idx  = np.where(y_train == 1)[0].tolist()
random.shuffle(benign_idx)
random.shuffle(malig_idx)

client_indices = []
sb = 0
sm = 0

for cid, (b_ct, m_ct) in enumerate(MANUAL_LABEL_ALLOC):
    idx = benign_idx[sb:sb + b_ct] + malig_idx[sm:sm + m_ct]
    random.shuffle(idx)
    client_indices.append(idx)
    sb += b_ct
    sm += m_ct

client_splits = [(X_train[idx], y_train[idx]) for idx in client_indices]

print("Actual per-client label counts (locked once):")
for cid, (_, yi) in enumerate(client_splits):
    total = len(yi)
    b = int(np.sum(yi == 0))
    m = int(np.sum(yi == 1))
    print(
        f"  Client {cid}: total={total:3d}, benign={b:3d} ({b/total:.1%}), malignant={m:3d} ({m/total:.1%})"
    )
print("\nClient splits are locked and reused for all experiments.\n")

# ============================================================
# 6. HELPER FUNCTIONS (DRIFT, METRICS, ENSEMBLE, TRAIN TABNET)
# ============================================================
def compute_client_drift(splits):
    """L2 distance between each client's mean and global mean."""
    all_X = np.vstack([Xi for Xi, _ in splits])
    global_mean = all_X.mean(axis=0)
    drifts_local = []
    for Xi, _ in splits:
        drifts_local.append(np.linalg.norm(Xi.mean(axis=0) - global_mean))
    return drifts_local

def score_vector(y_true, y_pred):
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
    rec  = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1   = f1_score(y_true, y_pred, average='macro', zero_division=0)
    return acc, prec, rec, f1

def ensemble_predict(models, sizes, X_eval, dp=False, dp_std=DP_AGG_NOISE_STD):
    total = float(sum(sizes))
    proba = np.zeros((X_eval.shape[0], 2), dtype=float)
    for clf, n_k in zip(models, sizes):
        w = n_k / total
        proba += w * clf.predict_proba(X_eval)
    if dp:
        proba += np.random.normal(0, dp_std, proba.shape)
        proba = np.clip(proba, 1e-6, 1-1e-6)
        proba = proba / proba.sum(axis=1, keepdims=True)
    return (proba[:, 1] >= 0.5).astype(int)

def train_tabnet(X_local, y_local, seed, mu=0.0):
    """
    Stable TabNet training with:
    - NO early stopping
    - NO batchnorm crashes
    - Fixed batch handling
    """

    torch.manual_seed(seed)
    np.random.seed(seed)

    local_gamma = 1.3 + mu

    clf = TabNetClassifier(
        **TABNET_PARAMS,
        gamma=local_gamma,
        seed=seed
    )

    n_samples = X_local.shape[0]

    # ---- SAFE BATCH SIZE ----
    # Never allow batch size < 4 (BatchNorm-safe)
    batch_size = max(4, min(32, n_samples))
    virtual_batch_size = min(4, batch_size)

    clf.fit(
        X_local,
        y_local,
        max_epochs=MAX_EPOCHS,     # FULL epochs, no early stop
        patience=0,                # ❌ disable early stopping
        batch_size=batch_size,
        virtual_batch_size=virtual_batch_size,
        num_workers=0,
        drop_last=True,            # prevents size=1 internal batches
        eval_set=None              # ❌ no validation = no early stop
    )

    return clf


# ============================================================
# 7. CLIENT DRIFT VISUALIZATION
# ============================================================
drifts = compute_client_drift(client_splits)
plt.figure(figsize=(7, 4))
plt.plot(range(N_CLIENTS), drifts, marker="o")
plt.xlabel("Client ID")
plt.ylabel("L2 distance from global mean")
plt.title("Client Drift Visualization (Feature Space)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# PART 1 complete. Run this cell, then request "Send Part 2".
# ============================================================
# TABNET + FEDERATED LEARNING — PART 2
# 50 Trials for 5 Experiments
# ============================================================

exp1_hist = []  # Central TabNet Baseline
exp2_hist = []  # FedAvg TabNet
exp3_hist = []  # FedAvg TabNet + DP
exp4_hist = []  # FedProx TabNet
exp5_hist = []  # FedProx TabNet + DP

print("========== STARTING 50 TRIALS ==========\n")

for t in range(1, N_TRIALS + 1):
    print(f"========== TRIAL {t}/{N_TRIALS} ==========\n")

    # -------------------------------------------------------
    # 1. Central TabNet Baseline
    # -------------------------------------------------------
    model1 = train_tabnet(X_train, y_train, seed=10000 + t, mu=0.0)
    y1 = model1.predict(X_test).reshape(-1)
    m1 = score_vector(y_test, y1)
    exp1_hist.append(m1)
    print(f"[1] Central TabNet -> Acc={m1[0]:.4f}, F1={m1[3]:.4f}")

    # -------------------------------------------------------
    # 2. FedAvg TabNet (No DP)
    # -------------------------------------------------------
    models2 = []
    sizes2  = []
    for cid, (Xi, yi) in enumerate(client_splits):
        X_loc = Xi.copy()
        if cid in FEATURE_SHIFT_CLIENTS:
            X_loc = X_loc + FEATURE_SHIFT_DELTA  # drift
        model2 = train_tabnet(X_loc, yi, seed=20000 + 10*t + cid, mu=0.0)
        models2.append(model2)
        sizes2.append(len(yi))
    y2 = ensemble_predict(models2, sizes2, X_test, dp=False)
    m2 = score_vector(y_test, y2)
    exp2_hist.append(m2)
    print(f"[2] FedAvg TabNet -> Acc={m2[0]:.4f}, F1={m2[3]:.4f}")

    # -------------------------------------------------------
    # 3. FedAvg TabNet + DP
    # -------------------------------------------------------
    y3 = ensemble_predict(models2, sizes2, X_test, dp=True)
    m3 = score_vector(y_test, y3)
    exp3_hist.append(m3)
    print(f"[3] FedAvg + DP -> Acc={m3[0]:.4f}, F1={m3[3]:.4f}")

    # -------------------------------------------------------
    # 4. FedProx TabNet (mu=0.2) — No DP
    # -------------------------------------------------------
    models4 = []
    sizes4  = []
    for cid, (Xi, yi) in enumerate(client_splits):
        X_loc = Xi.copy()
        if cid in FEATURE_SHIFT_CLIENTS:
            X_loc = X_loc + FEATURE_SHIFT_DELTA
        model4 = train_tabnet(X_loc, yi, seed=30000 + 10*t + cid, mu=0.2)
        models4.append(model4)
        sizes4.append(len(yi))
    y4 = ensemble_predict(models4, sizes4, X_test, dp=False)
    m4 = score_vector(y_test, y4)
    exp4_hist.append(m4)
    print(f"[4] FedProx TabNet -> Acc={m4[0]:.4f}, F1={m4[3]:.4f}")

    # -------------------------------------------------------
    # 5. FedProx TabNet + DP
    # -------------------------------------------------------
    y5 = ensemble_predict(models4, sizes4, X_test, dp=True)
    m5 = score_vector(y_test, y5)
    exp5_hist.append(m5)
    print(f"[5] FedProx + DP -> Acc={m5[0]:.4f}, F1={m5[3]:.4f}\n")

print("\n========== ALL 50 TRIALS COMPLETE ==========\n")

# ============================================================
# 8. METRIC SUMMARY (Mean ± STD)
# ============================================================

cols = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']
df1 = pd.DataFrame(exp1_hist, columns=cols)
df2 = pd.DataFrame(exp2_hist, columns=cols)
df3 = pd.DataFrame(exp3_hist, columns=cols)
df4 = pd.DataFrame(exp4_hist, columns=cols)
df5 = pd.DataFrame(exp5_hist, columns=cols)

summary_df = pd.DataFrame({
    "experiment": [
        "Central TabNet",
        "FedAvg",
        "FedAvg + DP",
        "FedProx",
        "FedProx + DP"
    ],
    "acc_mean":  [df['accuracy'].mean()        for df in [df1, df2, df3, df4, df5]],
    "acc_std":   [df['accuracy'].std()         for df in [df1, df2, df3, df4, df5]],
    "f1_mean":   [df['f1_macro'].mean()        for df in [df1, df2, df3, df4, df5]],
    "f1_std":    [df['f1_macro'].std()         for df in [df1, df2, df3, df4, df5]],
    "prec_mean": [df['precision_macro'].mean() for df in [df1, df2, df3, df4, df5]],
    "rec_mean":  [df['recall_macro'].mean()    for df in [df1, df2, df3, df4, df5]],
})

print("===== 50-TRIAL SUMMARY (MEAN ± STD) =====\n")
display(summary_df.round(4))

# ============================================================
# 9. PLOTS — Accuracy & F1 (mean over 50 trials)
# ============================================================

plt.figure(figsize=(9, 5))
plt.plot(summary_df['experiment'], summary_df['acc_mean'], marker='o')
plt.title("TabNet Accuracy Across Experiments")
plt.ylabel("Accuracy (mean)")
plt.xticks(rotation=20)
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(9, 5))
plt.plot(summary_df['experiment'], summary_df['f1_mean'], marker='o')
plt.title("TabNet F1 Score Across Experiments")
plt.ylabel("Macro F1 (mean)")
plt.xticks(rotation=20)
plt.grid(alpha=0.3)
plt.show()

print("\nPart 2 complete. Request:  **Send Part 3** (Statistical Tests).")
# ============================================================
# TABNET — PART 3
# FULL STATISTICAL TESTING BLOCK
# ============================================================

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel, wilcoxon, friedmanchisquare

# ------------------------------------------------------------
# Gather all experiment score arrays (Accuracy & F1)
# ------------------------------------------------------------
accs = {
    "Central": np.array(df1["accuracy"]),
    "FedAvg": np.array(df2["accuracy"]),
    "FedAvg+DP": np.array(df3["accuracy"]),
    "FedProx": np.array(df4["accuracy"]),
    "FedProx+DP": np.array(df5["accuracy"]),
}

f1s = {
    "Central": np.array(df1["f1_macro"]),
    "FedAvg": np.array(df2["f1_macro"]),
    "FedAvg+DP": np.array(df3["f1_macro"]),
    "FedProx": np.array(df4["f1_macro"]),
    "FedProx+DP": np.array(df5["f1_macro"]),
}

labels = list(accs.keys())
k = len(labels)

# ============================================================
# 1. Paired t-tests (Accuracy & F1)
# ============================================================
def paired_tests(metric_dict):
    results = pd.DataFrame(np.zeros((k, k)), index=labels, columns=labels)
    for a in labels:
        for b in labels:
            if a == b:
                results.loc[a, b] = np.nan
            else:
                t, p = ttest_rel(metric_dict[a], metric_dict[b])
                results.loc[a, b] = p
    return results

t_acc = paired_tests(accs)
t_f1  = paired_tests(f1s)

print("\n===== Paired t-test P-values (Accuracy) =====")
display(t_acc)

print("\n===== Paired t-test P-values (F1) =====")
display(t_f1)

# ============================================================
# 2. Wilcoxon Signed-Rank Tests (non-parametric)
# ============================================================
def wilcoxon_tests(metric_dict):
    results = pd.DataFrame(np.zeros((k, k)), index=labels, columns=labels)
    for a in labels:
        for b in labels:
            if a == b:
                results.loc[a, b] = np.nan
            else:
                try:
                    w, p = wilcoxon(metric_dict[a], metric_dict[b])
                except:
                    p = 1.0
                results.loc[a, b] = p
    return results

w_acc = wilcoxon_tests(accs)
w_f1  = wilcoxon_tests(f1s)

print("\n===== Wilcoxon P-values (Accuracy) =====")
display(w_acc)

print("\n===== Wilcoxon P-values (F1) =====")
display(w_f1)

# ============================================================
# 3. Friedman Test (Across all 5 methods)
# ============================================================
friedman_acc = friedmanchisquare(
    accs["Central"], accs["FedAvg"], accs["FedAvg+DP"], accs["FedProx"], accs["FedProx+DP"]
)

friedman_f1 = friedmanchisquare(
    f1s["Central"], f1s["FedAvg"], f1s["FedAvg+DP"], f1s["FedProx"], f1s["FedProx+DP"]
)

print("\n===== Friedman Test =====")
print(f"Accuracy: χ² = {friedman_acc[0]:.4f}, p = {friedman_acc[1]:.6f}")
print(f"F1 Score: χ² = {friedman_f1[0]:.4f}, p = {friedman_f1[1]:.6f}")

# ============================================================
# 4. Bonferroni Correction
# ============================================================
alpha = 0.05
comparisons = k * (k - 1) / 2
bonf_alpha = alpha / comparisons

print(f"\nBonferroni corrected alpha = {bonf_alpha:.5f}")

def apply_bonferroni(df):
    sig = df < bonf_alpha
    return sig.replace({True: "✓", False: "✗"})

bonf_t_acc = apply_bonferroni(t_acc)
bonf_w_acc = apply_bonferroni(w_acc)

bonf_t_f1 = apply_bonferroni(t_f1)
bonf_w_f1 = apply_bonferroni(w_f1)

print("\n===== Bonferroni Significance (t-test, Accuracy) =====")
display(bonf_t_acc)

print("\n===== Bonferroni Significance (Wilcoxon, Accuracy) =====")
display(bonf_w_acc)

print("\n===== Bonferroni Significance (t-test, F1) =====")
display(bonf_t_f1)

print("\n===== Bonferroni Significance (Wilcoxon, F1) =====")
display(bonf_w_f1)

# ============================================================
# 5. Heatmaps of p-values
# ============================================================
plt.figure(figsize=(10, 7))
sns.heatmap(t_acc, annot=True, cmap="viridis", fmt=".4f")
plt.title("Paired t-test p-values (Accuracy)")
plt.show()

plt.figure(figsize=(10, 7))
sns.heatmap(w_acc, annot=True, cmap="viridis", fmt=".4f")
plt.title("Wilcoxon p-values (Accuracy)")
plt.show()

plt.figure(figsize=(10, 7))
sns.heatmap(t_f1, annot=True, cmap="viridis", fmt=".4f")
plt.title("Paired t-test p-values (F1)")
plt.show()

plt.figure(figsize=(10, 7))
sns.heatmap(w_f1, annot=True, cmap="viridis", fmt=".4f")
plt.title("Wilcoxon p-values (F1)")
plt.show()

# ============================================================
# 6. Boxplots for Accuracy & F1 distributions
# ============================================================
acc_frame = pd.DataFrame(accs)
f1_frame  = pd.DataFrame(f1s)

plt.figure(figsize=(10, 6))
sns.boxplot(data=acc_frame)
plt.title("Accuracy Distribution Across Methods")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=f1_frame)
plt.title("F1 Score Distribution Across Methods")
plt.ylabel("Macro F1")
plt.xticks(rotation=20)
plt.show()

# ============================================================
# 7. SIGNIFICANCE SUMMARY TABLE (Final Table)
# ============================================================
summary_significance = pd.DataFrame(index=labels, columns=labels)

for i in labels:
    for j in labels:
        if i == j:
            summary_significance.loc[i, j] = "-"
        else:
            p = t_acc.loc[i, j]
            summary_significance.loc[i, j] = "✓" if p < bonf_alpha else "✗"

print("\n===== FINAL SIGNIFICANCE TABLE (Acc, Paired t-test + Bonferroni) =====")
display(summary_significance)

print("\n\n=== PART 3 COMPLETE — ALL STATISTICAL TESTS DONE===")
```
